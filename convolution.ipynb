### [Convolution](https://en.wikipedia.org/wiki/Convolution#Definition)



```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t=np.linspace(-5,5,1000)
w1=np.heaviside(t,1)-np.heaviside(t-4,1)
w2=np.heaviside(t,1)-np.heaviside(t-4,1)
convolution=np.convolve(w1,w2)
plt.plot(t,convolution, mode='same')
```
