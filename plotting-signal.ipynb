### Import library

```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
```

### Continuous Signal

Consider a CT signal $x(t)$. A real valued CT signal is plotted using
the standard ways to plot a function $x: R \rightarrow
R$. Consider the sinusoidal signal $x(t)=325 \sin(2\pi 50 t)$ for
a short interval of time we can plot this as:

```python
t = np.linspace(-0.02, 0.05, 1000)
fig, ax=plt.subplots(figsize=(15,5))
ax.plot(t, 325 * np.sin(2*np.pi*50*t), color='m')

ax.set_xlabel('t')
ax.set_ylabel('x(t)')
ax.set_title('Plot of CT signal $x(t)=325 \sin(2\pi 50 t)$')
ax.set_xlim([-0.02, 0.05])

# Turn on the minor TICKS, which are required for the minor GRID
#ax.minorticks_on()
#ax.grid(which='major', linestyle='-', linewidth='0.5', color='black')
#ax.grid(which='minor', linestyle='-', linewidth='0.5', color='0.65')
# Turn off the display of all ticks.
#ax.tick_params(which='both', # Options for both major and minor ticks
#                top=False, # turn off top ticks
#                left=False, # turn off left ticks
#                right=False,  # turn off right ticks
#                bottom=False) # turn off bottom ticks
ax.grid()
```

```python
t = np.linspace( 0, 1, 1000, endpoint = False )
x = np.cos( 2 * np.pi * 5 * t )
fig, ax=plt.subplots(figsize=(15,5))
ax.plot(t,x, color='b')

ax.set_xlabel( 't (second)' )
ax.set_ylabel( 'Amplitude' )
ax.set_xlim(left=0)

ax.grid()
```

### Discrete Signal

```python
n=np.arange(50)
dt = 0.05/50
x = np.sin(2 * np.pi * 50 * n * dt)
fig, ax=plt.subplots(figsize=(15,5))
ax.stem(n, x)
ax.set_xlabel('n')
ax.set_ylabel('x[n]')
ax.set_title('Plot of DT signal $x[n] = 325 \sin(2\pi 50 n \Delta t)$')
ax.grid()
```

### Plotting compex valued signal
Now consider the complex valued CT signal:
$$x(t) = e^{j 100 \pi t}$$
Plotting such a signal as if it were a real valued signal results in a
Python warning that complex numbers are casted to real by discarding
the imaginary parts. To see everything we have to plot both the real
and imaginary part of the signal.

```python
t = np.linspace(-0.02, 0.1, 1000)
fig, ax=plt.subplots(nrows=2, ncols=1,sharex=True,figsize=(15,5))

ax[0].plot(t, np.exp(2j*np.pi*50*t).real, color='m')
ax[0].set_xlabel('t')
ax[0].set_ylabel('x(t)')
ax[0].set_title('Real part of  $x(t)=e^{j 100 \pi t}$')
ax[0].set_xlim([-0.02, 0.1])

ax[1].plot(t, np.exp(2j*np.pi*50*t).imag, color='r')
ax[1].set_xlabel('t')
ax[1].set_ylabel('Re x(t)')
ax[1].set_title('Real part of  $x(t)=e^{j 100 \pi t}$')
ax[1].set_xlim([-0.02, 0.1])
```

Instead of plotting the real and imaginary part of a complex signal, we can also plot the magnitude and angle of the complex values.

```python
t = np.linspace(-0.02, 0.05, 1000)
fig, ax=plt.subplots(nrows=2, ncols=1,sharex=True,figsize=(15,5))
x=np.exp(2j*np.pi*50*t)
ax[0].plot(t, np.abs(x), color='m')
ax[0].set_xlabel('$t$')
ax[0].set_ylabel('$|x(t)|$')
ax[0].set_title('Absolute value  of  $x(t)=e^{j 100 \pi t}$')
ax[0].set_xlim([-0.02, 0.05])

ax[1].plot(t, np.angle(x), color='r')
ax[1].set_xlabel('$t$')
ax[1].set_ylabel('Angle of $x(t)$')
ax[1].set_title('Phase of  $x(t)=e^{j 100 \pi t}$')
ax[1].set_xlim([-0.02, 0.05])
```

Note that both plots of the complex signal are equivalent. They both
do represent the same signal. Also note the seemingly strange
behaviour of the phase (angle) plot. It looks quite discontinuous. But
it is really not because the phase jumps from +180 degrees to -180
degrees and that is of course the same angle! This phenomenon will be
seen quite a lot when plotting the phase of complex signals and
functions. It is called **phase wrapping**. If we would allow angles
outside the range from -180 to +180 we could obtain a perfectly
straight line (doing just that is called **phase unwrapping**).

### Square wave

```python
t = np.linspace(0, 1, 500, endpoint=False)
x=signal.square(2 * np.pi * 5 * t)
plt.figure(figsize=(15,5))
plt.plot(t,x, color='r')
plt.ylim(-2, 2)
plt.xlim(0,1)
```

### Sawtooth wave

```python
t = np.linspace(0, 1, 500, endpoint=False)
x=signal.sawtooth(2 * np.pi * 5 * t)
plt.figure(figsize=(15,5))
plt.plot(t,x, color='r')
plt.ylim(-2, 2)
plt.xlim(0,1)
```

### Step function

```python
t=np.linspace(-3,5,1000)
step=np.heaviside(t,1)
fig, ax=plt.subplots(figsize=(15,5))
ax.plot(t, step, color='r')
ax.set_title('Unit Step function')
ax.set_xlabel('time (s)')
ax.set_ylabel('Amplitude')

ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
```

```python
window=np.heaviside(t-1,1)-np.heaviside(t-3,1)
fig, ax=plt.subplots(figsize=(15,5))
ax.plot(t, window, color='r')
ax.set_title('Window function')
ax.set_xlabel('time (s)')
ax.set_ylabel('Amplitude')

ax.spines['left'].set_position('zero')
ax.spines['left'].set_color('blue')
ax.spines['bottom'].set_color('blue')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
```
